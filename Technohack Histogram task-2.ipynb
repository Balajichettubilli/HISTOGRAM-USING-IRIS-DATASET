```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load the iris dataset
data = pd.read_csv('iris.csv')
```

```python
# what is the shape our dataset[Rows & Columns] ?

print(data.shape)
```

```python
# What are the column names in our iris dataset ?

print(data.columns)
```

```python
 # How many Species in our dataset?
data["Species"].value_counts()
```

```python
# describe the dataset of iris
data.describe()
```

```python
# Histogram of Sepal length 

plt.figure(figsize = (10, 7))
x = data["SepalLengthCm"]

plt.hist(x, bins = 20, color = "yellow",rwidth=0.8,ec='blue')
plt.title("SEPAL LENGTH IN CM")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
plt.show()
```

```python
# Histogram of sepal width 

plt.figure(figsize = (10,7))
x = data['SepalWidthCm']

plt.hist(x, bins=20, color='orange', rwidth=0.8, ec='yellow')
plt.title('SEPAL WIDTH IN CM')
plt.xlabel('Sepal_Width_cm')
plt.ylabel('Count')
```

```python
# Histogram of Petal Length in Cm

plt.figure(figsize = (10,8))
x = data['PetalLengthCm']

plt.hist(x, bins=15, color='black',rwidth = 0.8, ec='red')
plt.title('PETAL LENGTH IN CM')
plt.xlabel('Petal_Length_cm')
plt.ylabel('Count')
```

```python
# Histogram of PetalLengthCm

plt.figure(figsize=(10,8))
x = data['PetalLengthCm']

plt.hist(x, bins=15, color='pink', rwidth=0.8, ec='yellow')
plt.title('PETAL LENGTH IN CM')
plt.xlabel('Petal_Length_Cm')
plt.ylabel('Count')
```
